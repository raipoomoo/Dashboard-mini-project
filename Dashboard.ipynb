# Dashboard for Teen Media Habits and Mental Health

### Background of Study

Teen mental health is an increasingly important issue in today's digital age. Factors such as social media usage, sleep habits, physical activity, and social interactions may influence teenagers' mental well-being. This study analyzes these factors to identify patterns related to mental health among teenagers using data visualization and analytics.

### Objective of Analysis

To analyze the factors that influence teenage mental health and identify how social media usage, sleep habits, physical activity, and social interactions are associated with mental health risk among teenagers.

### Loading the Raw Dataset

```python
import pandas as pd
df = pd.read_csv('Teen_Mental_Health.csv')
df.head()
```

#### Renaming Column

```python
df.columns = ["Age", "Gender", "Daily Social Media Hours", "Platform Usage", "Sleep Hours", "Screen Time Before Sleep", "Academic Performance", "Physical Activity", "Social Interaction Level", "Stress Level", "Anxiety Level", "Addiction Level", "Depression Label", "Mental Health Risk Score", "Sleep Quality", "Digital Wellbeing Flag"]
df["Gender"] = df["Gender"].str.capitalize()
df["Social Interaction Level"] = df["Social Interaction Level"].str.capitalize()
```

### Discussion

The analysis shows that social media usage, sleep patterns, physical activity, and social interaction are linked to teenagers’ mental health. Higher social media use and poor sleep are generally associated with higher stress, anxiety, and mental health risk, while healthier habits show better outcomes.

### Conclusion

This project analyzed factors affecting teenage mental health, including social media usage, sleep patterns, physical activity, and social interaction levels. The results show that lifestyle and digital habits are closely related to stress, anxiety, and overall mental health risk.

Overall, healthier habits such as better sleep, regular exercise, and balanced social media use are associated with improved mental well-being. The dashboard helps visualize these relationships clearly, although the results only show patterns and not direct causation.
